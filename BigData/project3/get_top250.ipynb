```python
from pykrx import stock
```

```python
kospi =  stock.get_market_ticker_list("20211130", market="KOSPI")
kosdaq =  stock.get_market_ticker_list("20211130", market="KOSPI")
```

```python
df = stock.get_market_cap("20211130")
print(df.head(250))
```

```python
import csv

corpcode = {}
to6digitnum = lambda x : "0"*(6-len(x))+x
to8digitnum = lambda x : "0"*(8-len(x))+x
with open('data/stockcode2corpcode.csv', newline='') as csvfile:
	spamreader = csv.reader(csvfile, delimiter=',')
	for i,row in enumerate(spamreader):
		if i != 0:
			corpcode[to6digitnum(row[2]) ] = [row[1],to8digitnum(row[0]),to6digitnum(row[2])]
```

```python
corpcode
```

```python
kospi_count = 0
kosdaq_count = 0

kospi_top250 = []
kosdaq_top250 = []

for d in df.index:
	if d in kospi:
		if kospi_count < 250 and d in corpcode and corpcode[d] not in kospi_top250:
			kospi_top250.append(corpcode[d])
			kospi_count += 1

	else:
		if kosdaq_count < 250  and d in corpcode and corpcode[d] not in kosdaq_top250:
			kosdaq_top250.append(corpcode[d])
			kosdaq_count += 1

	if kospi_count + kosdaq_count == 500:
		break





```

```python
kospi_top250
```

```python
wang_kospi = []
wang_kosdaq = []

with open('data/kospi_extend3.csv', newline='') as csvfile:
	spamreader = csv.reader(csvfile, delimiter=',')
	for i,row in enumerate(spamreader):
		if i != 0:
			wang_kospi.append(to8digitnum(row[2]))


with open('data/kodap_extend3.csv', newline='') as csvfile:
	spamreader = csv.reader(csvfile, delimiter=',')
	for i,row in enumerate(spamreader):
		if i != 0:
			wang_kosdaq.append(to8digitnum(row[2]))
```

```python
kospi_not_enclude = []
for i in kospi_top250:
	if i[1] not in wang_kospi:
		kospi_not_enclude.append(i)
```

```python
kosdaq_not_enclude = []
for i in kosdaq_top250:
	if i[1] not in wang_kosdaq:
		kosdaq_not_enclude.append(i)
```

```python
with open('kospi_top250.csv', 'w',encoding='utf-8-sig',newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',')
    for k in kospi_top250:
        spamwriter.writerow(k)
```
