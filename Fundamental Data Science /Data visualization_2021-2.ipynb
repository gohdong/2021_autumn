## 2. Scatter plot

```python
height=[165,177,160,180,185]
weight=[62,67,55,74,90]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(height,weight) # plt.scatter(x,y)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('H vs. W')
plt.grid(True)
plt.show()
```

```python
point_size=100
point_color='red'
marker_type='s'
trans=0.3 # 0부터 1사이 값
plt.scatter(height,weight,s=point_size,c=point_color,marker=marker_type,alpha=trans) # plt.scatter(x,y)
#plt.scatter(height,weight) # plt.scatter(x,y)
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('H vs. W')
plt.grid(True)
plt.show()
```

```python
# 마커 사이즈를 어떤 값을 바탕으로 조절하고, 마커에 텍스트 까지 넣는 경우
```

```python
city=['Seoul','Incheon','Gwangju','Daegu','Busan']
lat=[37.56,37.45,35.16,35.87,35.10]
lon=[126.97,126.7,126.85,128.6,129.07]
pop=[16154,2751,2995,128,4454]
```

```python
import numpy as np
pop_den=np.array(pop)*0.2
```

```python
plt.scatter(lon,lat,s=pop_den,alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Population')
for x,y,city_name in zip(lon,lat,city):
    plt.text(x,y,city_name)
plt.show()

#plt.text(x,y,'string')
```

## 3. Bar chart (x축 값이 범주형 변수)

```python
# Member_ID
# before diet (weight)
# after diet (weight)
```

```python
member_ID=['m_01','m_02','m_03']
before=[80,90,75]
after=[70,85,67]
```

```python
index=np.arange(1,4)
plt.bar(index,before,tick_label=member_ID,width=0.3) # plt.bar(x,y)
plt.show()
```

```python
index=np.arange(1,4)
plt.bar(index,before,tick_label=member_ID,width=0.3,label='before') # plt.bar(x,y)
plt.bar(index+0.3,after,tick_label=member_ID,width=0.3,label='after') # plt.bar(x,y)
plt.legend()
plt.show()
```

## 4. Pie chart

```python
fruit=['apple','banana','strawberry','orange']
pref=[7,6,3,2]
# plt.pie(freq)
```

```python
plt.pie(pref,labels=fruit,autopct='%0.2f%%')
# % 0.1f% % 소수점 첫째 자리까지 f:float %:percentage
plt.show()
```
